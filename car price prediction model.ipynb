# linear regression

```python
from sklearn import linear_model
```

```python
y= car_price.price
```

```python
car_price[car_price.isnull().any(axis=1)]
```

```python
pd.set_option('display.max_columns',100)
```

```python
car_price.cylindernumber.value_counts()
```

```python
cleanup_nums={'fueltype':{'gas':0,'diesel':1},'aspiration':{'std':0,'turbo':1},'doornumber':{'two':2,'four':4},'cylindernumber':{'four':4,'six':6,'five':5,'eight':8,'two':2,'twelve':12,'three':3}}
```

```python
cleanup_nums
```

```python
car_price.replace(cleanup_nums,inplace=True)
```

```python
car_price.dtypes
```

```python
car_price.columns
```

```python
x=car_price[['fueltype', 'aspiration',
       'doornumber', 'wheelbase',
       'carlength', 'carwidth', 'carheight', 'curbweight',
       'cylindernumber', 'enginesize', 'boreratio', 'stroke',
       'compressionratio', 'horsepower', 'peakrpm', 'citympg', 'highwaympg']]
```

```python
model=linear_model.LinearRegression()
```

```python
model.fit(x,y)
```

```python
y_predict=model.predict(x)
```

```python
y_predict
```

```python
predicted=pd.DataFrame(y_predict)
```

```python
predicted
```

```python
car_price['predicted']=predicted
```

```python
car_price
```

```python
model.score(x,y)
```

```python
car_price[['price','predicted']].plot()
```
